```python
import pandas as pd

def sliding_window(elements, window_size):
    if len(elements) <= window_size:
        return elements

    result = []
    for i in range(len(elements)- window_size + 1):
        result.append(elements[i:i+window_size])

    return pd.DataFrame(result)
```
