```python
import pandas as pd
import numpy as np
df= pd.read_csv('/content/new_df.csv', index_col=0,encoding='latin1')
#df1= pd.read_csv('/content/World-population-by-countries-dataset.csv')
df
```

```python
# Normalize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled= pd.DataFrame(df_scaled)
df_scaled
```

```python
from sklearn.model_selection import train_test_split
df_scaled= pd.read_csv('/content/df_scaled.csv')
# Separate the data into features (X) and target (y)
x = df_scaled.iloc[:, :30].values
Y = df_scaled.iloc[:, -5:].values
# Split the data into training and testing sets
x_train, x_test, Y_train, Y_test = train_test_split(x, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.metrics import mean_squared_error, explained_variance_score
# Create a linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(x_train, Y_train)

# Predict the carbon emissions for the test data
predictions = model.predict(x_test)
# Print the accuracy and MSE for each year in the test data
for i in range(5):
    year_prediction = predictions[0][i]
    year_actual = Y_test[0][i]
    if len(Y_test[:, i]) > 1:
        accuracy = r2_score(Y_test[:, i], predictions[:, i], multioutput='variance_weighted')
    else:
        accuracy = 0
    mse = mean_squared_error(Y_test[:, i], predictions[:, i])
    print(f"Year {i+1}: Prediction = {year_prediction:.2f}, Actual = {year_actual:.2f}, Accuracy = {accuracy:.2f}, MSE = {mse:.2f}")
```

```python
from sklearn.tree import DecisionTreeRegressor
# Create a decision tree regression model
model = DecisionTreeRegressor(random_state=42)

# Train the model on the training data
model.fit(x_train, Y_train)

# Predict the carbon emissions for the test data
predictions = model.predict(x_test)

# Print the accuracy and MSE for each year in the test data
for i in range(5):
    year_prediction = predictions[0][i]
    year_actual = Y_test[0][i]
    if len(Y_test[:, i]) > 1:
        accuracy = r2_score(Y_test[:, i], predictions[:, i], multioutput='variance_weighted')
    else:
        accuracy = 0
    mse = mean_squared_error(Y_test[:, i], predictions[:, i])
    print(f"Year {i+1}: Prediction = {year_prediction:.2f}, Actual = {year_actual:.2f}, Accuracy = {accuracy:.2f}, MSE = {mse:.2f}")
```

```python
# Create a random forest regression model with 100 trees
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model on the training data
rf.fit(x_train, Y_train)

# Predict the carbon emissions for the test data
y_pred = rf.predict(x_test)

# Print the accuracy and MSE for each year in the test data
for i in range(5):
    year_prediction = y_pred[0][i]
    year_actual = Y_test[0][i]
    if len(Y_test[:, i]) > 1:
        accuracy = r2_score(Y_test[:, i], y_pred[:, i], multioutput='variance_weighted')
    else:
        accuracy = 0
    mse = mean_squared_error(Y_test[:, i], y_pred[:, i])
    print(f"Year {i+1}: Prediction = {year_prediction:.2f}, Actual = {year_actual:.2f}, Accuracy = {accuracy:.2f}, MSE = {mse:.2f}")
```

```python

# Create a neural network model
from keras.models import Sequential
from keras.layers import Dense, Dropout
model = Sequential()
model.add(Dense(256, activation='relu', input_dim=30))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(5, activation='linear'))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse', 'mae'])

# Train the model on the training data
model.fit(x_train, Y_train, epochs=500, batch_size=64, validation_data=(x_test, Y_test))

# Predict the carbon emissions for the test data
predictions = model.predict(x_test)
# Print the accuracy and MSE for each year in the test data
for i in range(5):
    year_prediction = predictions[0][i]
    year_actual = Y_test[0][i]
    accuracy = r2_score(Y_test[:, i], predictions[:, i], multioutput='variance_weighted')
    mse = mean_squared_error(Y_test[:, i], predictions[:, i])
    print(f"Year {i+1}: Prediction = {year_prediction:.2f}, Actual = {year_actual:.2f}, Accuracy = {accuracy:.4f}, MSE = {mse:.4f}")
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
# Reshape the input data for LSTM
x_train = x_train.reshape((x_train.shape[0], 1, x_train.shape[1]))
x_test = x_test.reshape((x_test.shape[0], 1, x_test.shape[1]))

# Create a neural network model with LSTM
model = Sequential()
model.add(LSTM(256, activation='relu', input_shape=(1, 30)))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(5, activation='linear'))
# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse', 'mae'])
# Train the model on the training data
model.fit(x_train, Y_train, epochs=500, batch_size=64, validation_data=(x_test, Y_test))
# Predict the carbon emissions for the test data
predictions = model.predict(x_test)
# Print the accuracy and MSE for each year in the test data
for i in range(5):
    year_prediction = predictions[0][i]
    year_actual = Y_test[0][i]
    accuracy = r2_score(Y_test[:, i], predictions[:, i], multioutput='variance_weighted')
    mse = mean_squared_error(Y_test[:, i], predictions[:, i])
    print(f"Year {i+1}: Prediction = {year_prediction:.2f}, Actual = {year_actual:.2f}, Accuracy = {accuracy:.4f}, MSE = {mse:.4f}")
```
